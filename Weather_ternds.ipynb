```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# As I live in Hyderabad, I possess Hyderabad to compare with global temperatures.

```python
hdf = pd.read_csv('Hyderabad.csv')
```

```python
print('The shape of Hyderabad Dataframe is {}'.format(hdf.shape))
```

# Investigation of missing values.

```python
hdf.isna().sum()
```

```python
hdf['avg_temp'] = hdf['avg_temp'].fillna(hdf.avg_temp.mean())
```

```python
hdf.isnull().sum()
```

```python
hdf.info()
```

```python
gdf = pd.read_csv('Global_temp.csv')
```

```python
print('The shape of Global_temp Dataframe is {}'.format(gdf.shape))
```

```python
gdf.head()
```

```python
gdf.isna().sum()
```

```python
gdf.info()
```

```python
hyd_year = hdf['year']
hyd_temp = hdf['avg_temp']
global_year = gdf['year']
global_temp = gdf['avg_temp']
```

# Data Visualisation

```python
fig = plt.figure(figsize=(15,9))
plt.subplot(5,1,1)
plt.grid(True)
plt.plot(global_year,global_temp,label='Global Temprature over years')
plt.legend(loc=4)
plt.title('Global Temprature over years',fontsize=18)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Average Temp in Â°C',fontsize=15)
```

```python
fig = plt.figure(figsize=(15,9))
plt.subplot(5,1,1)
plt.grid(True)
plt.plot(hyd_year,hyd_temp,label='Hyderabad Temprature over years')
plt.legend(loc=4)
plt.title('Hyderabad Temprature over years',fontsize=20)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Average Temp in Â°C',fontsize=12)
```

```python
fig = plt.figure(figsize=(26,20))
plt.subplot(5,2,1)
plt.grid(True)
plt.plot(global_year,global_temp,label='Global Temprature over years')
plt.plot(hyd_year,hyd_temp,label='Hyderabad Temprature over years')
plt.legend(loc=6)
plt.ylim((0,30))
plt.title('Global and Hyderabad tempratures over years',fontsize=18)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Average Temp in Â°C',fontsize=15)
```

# Moving average for 1 year

```python
fig = plt.figure(figsize=(26,20))
window_state = 1
hyd_mvg = pd.DataFrame(hyd_temp.rolling(window=window_state).mean())
global_mvg = pd.DataFrame(global_temp.rolling(window=window_state).mean())
plt.subplot(5,2,2)
plt.grid()
plt.plot(global_year,global_mvg,label='Global Temperature')
plt.plot(hyd_year,hyd_mvg,label='Hyderabad Temprature')
plt.legend(loc=4)
plt.ylim((0,30))
plt.title('Moving average for 1 year',fontsize=18)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Average Temp in Â°C',fontsize=15)
```

# Moving average for 3 years

```python
fig = plt.figure(figsize=(26,20))
window_state = 3
hyd_mvg = pd.DataFrame(hyd_temp.rolling(window=window_state).mean())
global_mvg = pd.DataFrame(global_temp.rolling(window=window_state).mean())
plt.subplot(5,2,2)
plt.grid()
plt.plot(global_year,global_mvg,label='Global Temprature')
plt.plot(hyd_year,hyd_mvg,label='Hyderabad Temprature')
plt.legend(loc=4)
plt.ylim((0,30))
plt.title('Moving average temperatures for 3 years',fontsize=18)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Average Temp in Â°C',fontsize=15)
```

# Moving average for 7 years

```python
fig = plt.figure(figsize=(26,20))

window_state = 7 

hyd_mvg = pd.DataFrame(hyd_temp.rolling(window=window_state).mean())
global_mvg = pd.DataFrame(global_temp.rolling(window=window_state).mean())

plt.subplot(5,2,2)
plt.grid()
plt.plot(global_year,global_mvg,label=' Global Temprature')
plt.plot(hyd_year,hyd_mvg,label='Hyderabad Temprature')
plt.legend(loc=4)
plt.ylim((0,30))
plt.title('Moving average for 7 years',fontsize=18)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Average Temp in Â°C',fontsize=15)
```

# Moving average for 10 years

```python
fig = plt.figure(figsize=(26,20))

window_state = 10

hyd_mvg = pd.DataFrame(hyd_temp.rolling(window=window_state).mean())
global_mvg = pd.DataFrame(global_temp.rolling(window=window_state).mean())

plt.subplot(5,2,2)
plt.grid()
plt.plot(global_year,global_mvg,label='Global Temprature')
plt.plot(hyd_year,hyd_mvg,label='Hyderabad Temprature')
plt.legend(loc=4)
plt.ylim((0,30))
plt.title('Moving average for 10 years',fontsize=18)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Average Temp in Â°C',fontsize=15)
```

```python
mgl = np.max(global_temp)
mngl = np.min(global_temp)
ygl = max(global_year)
ymgl = min(global_year)
gavg = np.mean(global_temp)
print('The global minimum temperature from {} to {} is {}'.format(ymgl,ygl,mngl))
print('The global maximum temperature from {} to {} is {}'.format(ymgl,ygl,mgl))
```

```python
hgl = np.max(hyd_temp)
hngl = np.min(hyd_temp)
yhl = max(hyd_year)
ymhl = min(hyd_year)
havg = np.mean(hyd_temp)
print('The Hyderabad minimum temperature from {} to {} is {}'.format(ymhl,yhl,hngl))
print('The Hyderabad mamximum temperature from {} to {} is {}'.format(ymhl,yhl,hgl))
```

```python
print('The Average Temperature Difference Between Global and Hyderabad is {:.2f}'.format(havg-gavg))
```
